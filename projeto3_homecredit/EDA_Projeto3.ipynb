### Gráficos e outros insights

```python
def plot_stats(feature, label_rotation=False, horizontal_layout=True):
    """
    Função para visualizar a distribuição de uma variável categórica e sua relação com o TARGET.

    Parâmetros:
    - feature (str): Nome da coluna do DataFrame df_train a ser analisada.
    - label_rotation (bool): Se True, rotaciona os rótulos do eixo X para melhor visualização.
    - horizontal_layout (bool): Se True, plota os gráficos lado a lado; caso contrário, empilha os gráficos.
    """

    # Conta a frequência de cada categoria na variável escolhida
    temp = df_train[feature].value_counts()

    # Cria um DataFrame para armazenar as contagens
    df1 = pd.DataFrame({feature: temp.index, 'Número de contratos': temp.values})

    # Calcula a média do TARGET para cada categoria da variável
    cat_perc = df_train[[feature, 'TARGET']].groupby([feature], as_index=False).mean()

    # Ordena as categorias pelo valor do TARGET, do maior para o menor
    cat_perc.sort_values(by='TARGET', ascending=False, inplace=True)

    # Define o layout dos gráficos (lado a lado ou empilhados)
    if horizontal_layout:
        fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))
    else:
        fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(12,14))

    # Define o estilo do gráfico
    sns.set_color_codes("pastel")

    # Plota o gráfico de barras com a contagem de cada categoria
    s = sns.barplot(ax=ax1, x=feature, y="Número de contratos", data=df1)

    # Rotaciona os rótulos do eixo X se necessário
    if label_rotation:
        s.set_xticklabels(s.get_xticklabels(), rotation=90)

    # Plota o gráfico de barras com a porcentagem de TARGET=1 para cada categoria
    s = sns.barplot(ax=ax2, x=feature, y='TARGET', order=cat_perc[feature], data=cat_perc)

    # Rotaciona os rótulos do eixo X se necessário
    if label_rotation:
        s.set_xticklabels(s.get_xticklabels(), rotation=90)

    # Define o rótulo do eixo Y do segundo gráfico
    plt.ylabel('Percentual de TARGET=1 [%]', fontsize=10)

    # Ajusta o tamanho das fontes dos eixos
    plt.tick_params(axis='both', which='major', labelsize=10)

    # Exibe os gráficos
    plt.show()
```

```python
def plot_feature_distribution(df, feature, target_col='TARGET'):
    """
    Plota a distribuição de uma variável contínua diferenciando os valores de TARGET.

    Parâmetros:
        df: DataFrame contendo os dados
        feature: Nome da variável a ser analisada
        target_col: Nome da variável-alvo (TARGET)
    """
    plt.figure(figsize=(10,6))
    plt.title(f"Distribuição de {feature} por {target_col}")

    # Separando por TARGET 0 e 1
    sns.kdeplot(df[df[target_col] == 0][feature], label="TARGET = 0", shade=True)
    sns.kdeplot(df[df[target_col] == 1][feature], label="TARGET = 1", shade=True)

    plt.xlabel(feature)
    plt.ylabel("Densidade")
    plt.legend()
    plt.show()
```

### Correlações

```python
# Selecionar apenas colunas numéricas
df_numeric = df_train.select_dtypes(include=[np.number])

# Calcular a matriz de correlação ignorando valores nulos
correlation_matrix = df_numeric.corr()

# Criar um heatmap da matriz de correlação
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False, fmt=".2f", linewidths=0.5)
plt.title("Heatmap das Correlações")
plt.show()
```

```python
# Transformar a matriz de correlação em um DataFrame para ordenação
correlation_pairs = correlation_matrix.unstack().reset_index()
correlation_pairs.columns = ['Feature 1', 'Feature 2', 'Correlation']

# Remover correlações de uma variável com ela mesma (valor = 1)
correlation_pairs = correlation_pairs[correlation_pairs['Feature 1'] != correlation_pairs['Feature 2']]

# Remover valores duplicados (pois a matriz de correlação é simétrica)
correlation_pairs['sorted_pair'] = correlation_pairs.apply(lambda row: tuple(sorted([row['Feature 1'], row['Feature 2']])), axis=1)
correlation_pairs = correlation_pairs.drop_duplicates(subset=['sorted_pair']).drop(columns=['sorted_pair'])

# Ordenar pelas correlações absolutas para destacar os maiores valores
top_10_correlations = correlation_pairs.sort_values(by='Correlation', ascending=False).head(10)
bottom_10_correlations = correlation_pairs.sort_values(by='Correlation', ascending=True).head(10)

# Exibir as 10 maiores correlações
print("Top 10 maiores correlações:")
print(top_10_correlations)

# Exibir as 10 menores correlações
print("\nTop 10 menores correlações:")
print(bottom_10_correlations)
```

### Remoções - Pré-processamentos - Visualização de Outliers

```python
# Definir um threshold s
threshold = 0.8

# Selecionar apenas colunas numéricas
numeric_df = df_train.select_dtypes(include=['number']).dropna()

# Calcular a matriz de correlação absoluta
corr_matrix = numeric_df.corr().abs()

# Criar uma máscara para manter apenas a parte superior da matriz de correlação
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Selecionar colunas com correlação acima do threshold
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

print(f'Total de colunas a serem possivelmente removidas: {len(to_drop)}')
print(to_drop)
```

Esse código realiza a remoção de colunas altamente correlacionadas para evitar redundâncias no conjunto de dados

```python
# Correlação das colunas a serem removidas com o TARGET
corr_with_target = df_train[to_drop].corrwith(df_train['TARGET']).abs().sort_values(ascending=False)

# Mostrar as 10 variáveis mais correlacionadas com TARGET
print(corr_with_target.head(10))
```

```python
# Lista das 10 colunas mais correlacionadas com TARGET que devemos manter
cols_to_keep = [
    'REGION_RATING_CLIENT_W_CITY', 'FLAG_EMP_PHONE', 'FLOORSMAX_MEDI',
    'FLOORSMAX_MODE', 'AMT_GOODS_PRICE', 'ELEVATORS_AVG', 'ELEVATORS_MEDI',
    'FLOORSMIN_MEDI', 'LIVINGAREA_AVG', 'LIVINGAREA_MEDI'
]

# Lista final de colunas para remover (as que estavam em to_drop, mas não estão em cols_to_keep)
final_to_drop = [col for col in to_drop if col not in cols_to_keep]

# Remover as colunas pouco relevantes
df_train = df_train.drop(columns=final_to_drop)

print(f'Total de colunas removidas: {len(final_to_drop)}')
print(f'Novo formato do dataframe: {df_train.shape}')
```

```python
# Selecionar algumas colunas numéricas para visualizar outliers
num_cols = ['AMT_CREDIT', 'AMT_INCOME_TOTAL', 'AMT_ANNUITY', 'DAYS_BIRTH', 'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']

# Criar boxplots para cada variável selecionada
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols):
    plt.subplot(2, 4, i + 1)  # Organiza os gráficos em 2 linhas e 4 colunas
    sns.boxplot(x=df_train[col])
    plt.title(f"Boxplot de {col}")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Para evitar imputações artificiais que podem gerar data leakage ou introduzir ruído desnecessário no modelo, optei por dropar as colunas cujo valores nulos são muito altos (mais de 65%) .

```python
# Lista das colunas a serem removidas
colunas_para_dropar = [
    "COMMONAREA_AVG", "NONLIVINGAPARTMENTS_AVG", "FONDKAPREMONT_MODE",
    "FLOORSMIN_AVG", "FLOORSMIN_MEDI", "YEARS_BUILD_AVG", "OWN_CAR_AGE"
]

# Remover do dataset
df_train = df_train.drop(columns=colunas_para_dropar)

print(f"Total de colunas removidas: {len(colunas_para_dropar)}")
print(f"Novo formato do dataframe: {df_train.shape}")
```

```python
from sklearn.impute import SimpleImputer

# Recalcular colunas categóricas e numéricas que ainda possuem valores nulos
categorical_vars = [var for var in df_train.columns if df_train[var].isnull().mean() > 0 and df_train[var].dtypes == 'O']
numeric_vars = [var for var in df_train.columns if var not in categorical_vars and df_train[var].isnull().mean() > 0]

# Criar imputador para colunas numéricas (preencher com a mediana)
if numeric_vars:  # Evita erro caso não haja colunas numéricas com nulos
    imputer_num = SimpleImputer(strategy='median')
    df_train[numeric_vars] = imputer_num.fit_transform(df_train[numeric_vars])

# Criar imputador para colunas categóricas (preencher com a moda)
if categorical_vars:  # Evita erro caso não haja colunas categóricas com nulos
    imputer_cat = SimpleImputer(strategy='most_frequent')
    df_train[categorical_vars] = imputer_cat.fit_transform(df_train[categorical_vars])

# Verificar se ainda existem valores nulos
print("Valores nulos após a imputação:", df_train.isnull().sum().sum())  # Deve ser 0 ou muito próximo
```
